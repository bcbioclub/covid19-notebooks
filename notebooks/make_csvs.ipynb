```python
import pickle
import os
import csv
```

```python
no_mutants_found = ['NC_045512.2']
```

```python
amino_acid_names = {
  'T': 'Threonine',
  'Q': 'Glutamine',
  'S': 'Serine',
  'N': 'Asparagine',
  'F': 'Phenylalanine',
  'M': 'Methionine',
  'L': 'Leucine',
  'V': 'Valine',
  'W': 'Tryptophan',
  'A': 'Alanine',
  'I': 'Isoleucine',
  'E': 'Glutamic acid',
  'D': 'Aspartic Acid',
  'K': 'Lysine',
  'R': 'Arginine',
  'C': 'Cysteine',
  'G': 'Glycine',
  'P': 'Proline',
  'H': 'Histidine',
  'Y': 'Tyrosine'
}
```

```python
header = [
  'strain',
  'protein id',
  'product',
  'gene',
  'type',
  'feat start',
  'feat end',
  'position',
  'org',
  'new',
  'is color same',
  'org color',
  'new color',
  'is conserved',
  'org type',
  'new type'
]

pickles_path = './drive/MyDrive/Colab Notebooks/pickles/'
csvs_path = './drive/MyDrive/Colab Notebooks/csvs/'
csv_file = 'bio_club_mutations_data.csv'
filepath = csvs_path + csv_file
with open(filepath, 'w', newline='') as f:
  writer = csv.writer(f)
  writer.writerow(header)
  for file_name in os.listdir(pickles_path):
    if '.pickle' not in file_name:
      continue
    comparison_strain = file_name.replace('.pickle', '')
    if comparison_strain in no_mutants_found:
      continue
    strain = comparison_strain
    pickles_path = './drive/MyDrive/Colab Notebooks/pickles/'
    pickles_file = strain + '.pickle'
    filepath = pickles_path + pickles_file
    objects = []
    with (open(filepath, "rb")) as openfile:
      while True:
        try:
          objects.append(pickle.load(openfile))
        except EOFError:
          break
    comparison = objects[0]
    for feature in comparison:
      for mutant in feature['mutants']:
        org = amino_acid_names[mutant[1]] if mutant[1] != 'X' else 'unknown'
        new = amino_acid_names[mutant[2]] if mutant[2] != 'X' else 'unknown'
        if 'unknown' in [org, new]:
          continue
        org_color = None
        new_color = None
        if ' to ' in mutant[3]:
          org_color, new_color = mutant[3].split(' to ')
        org_type = None
        new_type = None
        if ' to ' in mutant[4]:
          org_type, new_type = mutant[4].split(' to ')
        row = [
          strain,
          feature.get('protein_id'),
          feature.get('product'),
          feature.get('gene'),
          feature.get('type'),
          feature.get('start'),
          feature.get('end'),
          mutant[0],
          org,
          new,
          True if mutant[3] == 'same' else False,
          org_color,
          new_color,
          True if mutant[4] == 'conservative' else False,
          org_type,
          new_type
        ]
        writer.writerow(row)
```
